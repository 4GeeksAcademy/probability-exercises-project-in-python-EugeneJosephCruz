## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import random
def experiment(num_trials):
    count = 0
    for i in range(num_trials):
        dice1 = random.randint(1,6)
        dice2 = random.randint(1,6)
        if (dice1 + dice2 > 7) or ((dice1 + dice2)%2==0):
            count += 1
    return count

print(f"Estimated probability of finding that the total score is even or greater than 7 for two dice rolls: {experiment(1000)/10}%")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import pandas as pd
import numpy as np
import random 

ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"
#df = pd.Series(ball_box)
print(ball_box)
#print(df)
            
# TODO
balls_picked = ["1st","2nd","3rd","4th","5th"]
count_red2_white3=0
count_all_same_color=0

for _ in range(1000):
    for i in range(5):
        balls_picked[i]=random.choice(ball_box)
        
    #print("balls_picked: ",balls_picked)
    red_balls_picked = balls_picked.count('Red')
    white_balls_picked = balls_picked.count('White')
    green_balls_picked = balls_picked.count('Green')
    #print("red_balls_picked: ", red_balls_picked)
    #print("white_balls_picked: ", white_balls_picked)
    #print("green_balls_picked: ", green_balls_picked)
    if (red_balls_picked == 2) and (white_balls_picked==3):
        count_red2_white3+=1
        #print("red balls are 2 and white balls are 3")
    elif (red_balls_picked==5)or(white_balls_picked==5)or(green_balls_picked==5):
        count_all_same_color+=1
        #print("all the balls are the same color")  

print("The probability that we take 3 white and 2 red after 1000 trials: ",count_red2_white3/10)
print("The probability that all are the same color after 1000 trials: ", count_all_same_color/10)

```
